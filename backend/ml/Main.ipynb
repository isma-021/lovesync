```python
import pandas as pd
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
users=pd.read_csv('dataset.csv')
```

```python
 users.head(10)
```

```python
users.columns
```

```python
users=users[['id', 'Nombre y Apellido', 'Intereses', 'Bio (Basada en referencias)']]
```

```python
users['name'] = users['Nombre y Apellido']
```

```python
users['bio'] = users['Bio (Basada en referencias)']
```

```python
users['int'] = users['Intereses']
```

```python
users['intdo'] = users['int'] + users['bio']
```

```python
users
```

```python
users_new = users.drop(columns=['Nombre y Apellido', 'Bio (Basada en referencias)', 'Intereses', 'int', 'bio'])
```

```python
users_new
```

```python
stop_words_es = ['el', 'la', 'los', 'las', 'de', 'y', 'a', 'en']
```

```python
cv=CountVectorizer(max_features=100, stop_words=stop_words_es)
```

```python
vector=cv.fit_transform(users_new['intdo'].values.astype('U')).toarray()
```

```python
vector.shape
```

```python
similaridad=cosine_similarity(vector)
```

```python
similaridad
```

```python
def recomendacion(persona):
    indexi=users_new[users_new['name']==persona].index[0]
    distance = sorted(list(enumerate(similaridad[indexi])), reverse=True, key=lambda vector:vector[1])
    for i in distance[1:5]:
        print(users_new.iloc[i[0]]['name'])
```

```python
recomendacion("Zoe")
```
